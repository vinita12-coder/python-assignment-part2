# ***Task 1 — Explore the Menu***

```python
menu = {
    "Paneer Tikka":   {"category": "Starters",  "price": 180.0, "available": True},
    "Chicken Wings":  {"category": "Starters",  "price": 220.0, "available": False},
    "Veg Soup":       {"category": "Starters",  "price": 120.0, "available": True},
    "Butter Chicken": {"category": "Mains",     "price": 320.0, "available": True},
    "Dal Tadka":      {"category": "Mains",     "price": 180.0, "available": True},
    "Veg Biryani":    {"category": "Mains",     "price": 250.0, "available": True},
    "Garlic Naan":    {"category": "Mains",     "price":  40.0, "available": True},
    "Gulab Jamun":    {"category": "Desserts",  "price":  90.0, "available": True},
    "Rasgulla":       {"category": "Desserts",  "price":  80.0, "available": True},
    "Ice Cream":      {"category": "Desserts",  "price": 110.0, "available": False},
}

# --- Part 1: Print menu category wise ---

categories = ["Starters", "Mains", "Desserts"]
for category in categories:
    print(f"\n===== {category} =====")

#finding item & check availability
    for item_name, details in menu.items():
        if details["category"] == category:
            availability = "[Available]" if details["available"] else "[Unavailable]"
            print(f"{item_name:<16} ₹{details['price']:<8.2f} {availability}")

# --- Part 2: Stats using dictionary methods ---

all_items     = list(menu.keys())
total_items   = len(all_items)

# count items where available == True
available_items = sum(1 for v in menu.values() if v["available"])

# most expensive — find item with max price
most_expensive = max(menu, key=lambda item: menu[item]["price"])

# items priced under 150
cheap_items = {name: details for name, details in menu.items() if details["price"] < 150}

print(f"\nTotal items     : {total_items}")
print(f"Available items : {available_items}")
print(f"Most expensive  : {most_expensive} (₹{menu[most_expensive]['price']})")

print("\nItems under ₹150:")
for name, details in cheap_items.items():
    print(f"  {name} — ₹{details['price']}")
```

# ***Task 2: Cart Operations***

```python
cart = []  #Createing a empty list

# ---  add item to cart ---
def add_item(item_name, quantity):
    # Check if item exists in menu
    if item_name not in menu:
        print(f"'{item_name}' does not exist in menu.")
        return

    # Checking if item is available
    if not menu[item_name]["available"]:
        print(f"'{item_name}' is currently unavailable.")
        return

    # Checking if item already in cart — if yes, then update quantity
    for entry in cart:
        if entry["item"] == item_name:
            entry["quantity"] += quantity
            print(f"Updated '{item_name}' quantity to {entry['quantity']}")
            return

    # Item is not in cart then adding new entry
    cart.append({
        "item": item_name,
        "quantity": quantity,
        "price": menu[item_name]["price"]
    })
    print(f"Added '{item_name}' x{quantity}")


# ---  remove item from cart ---
def remove_item(item_name):
    for entry in cart:
        if entry["item"] == item_name:
            cart.remove(entry)
            print(f"Removed '{item_name}' from cart.")
            return
    print(f"'{item_name}' is not in the cart.")


# ---  To update quantity ---
def update_quantity(item_name, new_qty):
    for entry in cart:
        if entry["item"] == item_name:
            entry["quantity"] = new_qty
            print(f"Updated '{item_name}' quantity to {new_qty}")
            return
    print(f"'{item_name}' is not in the cart.")


# --- Simulating ---

add_item("Paneer Tikka", 2)
print("Cart:", cart)

add_item("Gulab Jamun", 1)
print("Cart:", cart)

add_item("Paneer Tikka", 1)
print("Cart:", cart)

add_item("Mystery Burger", 1)
print("Cart:", cart)

add_item("Chicken Wings", 1)
print("Cart:", cart)

remove_item("Gulab Jamun")
print("Cart:", cart)



 #--- Final Order Summary ---

print("\n========== Order Summary ==========")

subtotal = 0
for entry in cart:
    line_total = entry["price"] * entry["quantity"]
    subtotal += line_total
    print(f"{entry['item']:<20} x{entry['quantity']}    ₹{line_total:.2f}")

gst        = round(subtotal * 0.05, 2)
total_pay  = round(subtotal + gst, 2)

print("------------------------------------")
print(f"{'Subtotal:':<28} ₹{subtotal:.2f}")
print(f"{'GST (5%):':<28} ₹{gst:.2f}")
print(f"{'Total Payable:':<28} ₹{total_pay:.2f}")
print("====================================")
```

# ***Task 3:- Inventory + Deep Copy***

```python
import copy

inventory = {
    "Paneer Tikka":   {"stock": 10, "reorder_level": 3},
    "Chicken Wings":  {"stock":  8, "reorder_level": 2},
    "Veg Soup":       {"stock": 15, "reorder_level": 5},
    "Butter Chicken": {"stock": 12, "reorder_level": 4},
    "Dal Tadka":      {"stock": 20, "reorder_level": 5},
    "Veg Biryani":    {"stock":  6, "reorder_level": 3},
    "Garlic Naan":    {"stock": 30, "reorder_level": 10},
    "Gulab Jamun":    {"stock":  5, "reorder_level": 2},
    "Rasgulla":       {"stock":  4, "reorder_level": 3},
    "Ice Cream":      {"stock":  7, "reorder_level": 4},
}

# ---  Deep copying  before making any changes ---
inventory_backup = copy.deepcopy(inventory)

# Demonstrate deep copy works — change inventory, backup stays same
inventory["Paneer Tikka"]["stock"] = 60
print("After change:")
print(f"inventory stock      : {inventory['Paneer Tikka']['stock']}")       # 999
print(f"inventory_backup stock: {inventory_backup['Paneer Tikka']['stock']}") # still 10

# Restore inventory back to original
inventory["Paneer Tikka"]["stock"] = 10
print("Inventory restored.\n")


#--- Checking the stocks after task 2---

for entry in cart:
    item_name = entry["item"]
    qty_needed = entry["quantity"]

    current_stock = inventory[item_name]["stock"]

    if qty_needed > current_stock:
        print(f" Warning: Only {current_stock} unit(s) of '{item_name}' available. Deducting {current_stock}.")
        inventory[item_name]["stock"] = 0
    else:
        inventory[item_name]["stock"] -= qty_needed
        print(f"Deducted {qty_needed} unit(s) of '{item_name}'. Remaining: {inventory[item_name]['stock']}")


# ---  Reorder alerts ---
print("\n--- Reorder Alerts ---")
for item_name, details in inventory.items():
    if details["stock"] <= details["reorder_level"]:
        print(f" Reorder Alert: {item_name} — Only {details['stock']} unit(s) left (reorder level: {details['reorder_level']})")


# --- Print both to confirm deep copy protected backup ---
print("\n--- Current Inventory (after deductions) ---")
for name, d in inventory.items():
    print(f"{name:<16}: stock = {d['stock']}")

print("\n--- Backup Inventory (unchanged) ---")
for name, d in inventory_backup.items():
    print(f"{name:<16}: stock = {d['stock']}")
```

# ***Task 4- Daily Sales Log Analysis***

```python
# --- Sales log data  ---

sales_log = {
    "2025-01-01": [
        {"order_id": 1,  "items": ["Paneer Tikka", "Garlic Naan"],          "total": 220.0},
        {"order_id": 2,  "items": ["Gulab Jamun", "Veg Soup"],              "total": 210.0},
        {"order_id": 3,  "items": ["Butter Chicken", "Garlic Naan"],        "total": 360.0},
    ],
    "2025-01-02": [
        {"order_id": 4,  "items": ["Dal Tadka", "Garlic Naan"],             "total": 220.0},
        {"order_id": 5,  "items": ["Veg Biryani", "Gulab Jamun"],           "total": 340.0},
    ],
    "2025-01-03": [
        {"order_id": 6,  "items": ["Paneer Tikka", "Rasgulla"],             "total": 260.0},
        {"order_id": 7,  "items": ["Butter Chicken", "Veg Biryani"],        "total": 570.0},
        {"order_id": 8,  "items": ["Garlic Naan", "Gulab Jamun"],           "total": 130.0},
    ],
    "2025-01-04": [
        {"order_id": 9,  "items": ["Dal Tadka", "Garlic Naan", "Rasgulla"], "total": 300.0},
        {"order_id": 10, "items": ["Paneer Tikka", "Gulab Jamun"],          "total": 270.0},
    ],
}


# --- Revenue per day ---

def print_revenue_table(log):
    print("\nDate         | Revenue")
    print("-" * 25)
    for date, orders in log.items():
        daily_total = sum(order["total"] for order in orders)  # sum all order totals for that day
        print(f"{date}  | ₹{daily_total:.2f}")


print_revenue_table(sales_log)


# ---  Best selling day ---

def best_day(log):
    best = max(log, key=lambda date: sum(o["total"] for o in log[date]))
    best_revenue = sum(o["total"] for o in log[best])
    print(f"\nBest selling day: {best} (₹{best_revenue:.2f})")


best_day(sales_log)


# ---  Most ordered item ---

item_count = {}  # dictionary to track how many orders each item appears in

for date, orders in sales_log.items():
    for order in orders:
        for item in order["items"]:
            if item not in item_count:
                item_count[item] = 0
            item_count[item] += 1

most_ordered = max(item_count, key=lambda x: item_count[x])
print(f"\nMost ordered item: {most_ordered} ({item_count[most_ordered]} orders)")


# ---  Add new day and reprint ---

sales_log["2025-01-05"] = [
    {"order_id": 11, "items": ["Butter Chicken", "Gulab Jamun", "Garlic Naan"], "total": 490.0},
    {"order_id": 12, "items": ["Paneer Tikka", "Rasgulla"],                     "total": 260.0},
]

print("\n--- Updated Revenue Table ---")
print_revenue_table(sales_log)
best_day(sales_log)


# --- Numbered list of all orders using enumerate ---

print("\n--- All Orders ---")
counter = 1  #

for date, orders in sales_log.items():
    for order in orders:
        items_str = ", ".join(order["items"])
        print(f"{counter}. [{date}] Order #{order['order_id']}  — ₹{order['total']:.2f} — Items: {items_str}")
        counter += 1
```
